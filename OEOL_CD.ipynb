```python
def obj(x, y):
  result = (x-2)**2 + (y+3)**2 + x*y
  return result

x, y = 0.0, 0.0
obj_old = obj(x, y)
for i in range(100):
  x = 2 - y/2
  y = -(x/2 + 3)

  obj_new = obj(x, y)
  print(f"Iter {i+1:2d}: x = {x:.4f}, y = {y:.4f}, f = {obj_new:.4f}")

  if abs(obj_new - obj_old) < 0.01:
    print(f"\nConverge after {i+1} iterations!")
    break

  obj_old = obj_new

print(f"\nFinal result: x = {x:.4f}, y = {y:.4f}, f = {obj_new:.4f}")
```
